```python
# even odd count in series
#l=[1,2,3,4,5,6,7,7,8,9,11,14,25,36,52,20,3]
Number = int(input("\nPlease Enter the Range : "))
count_even=count_odd=0
for i in range(1,Number):
#for i in l:
    print(i)
    if i%2==0:
        count_even+=1 
    elif i%2!=0:
        count_odd+=1
print('Even count numbers in series is',count_even)
print('Odd count numbers in series is',count_odd)
```
